## ML models

```python
# Now, we already have some ideas on 'what kind of customers/behaviors have higher response to the marketing promotions'
# e.g. ages from 20 to 40 and from 70 to 90  have higher response.
#      customers who are alone with PhD education level and absurd with graduated education level,and
#      those are widow with a master-degree and yolo with PhD education level have higher response.
# ...
# These three are segmentation level
# We do predictions on individual level, build a model to predict the probability of higher response rate.
```

```python
# 4.1 do data preprocessing
```

```python
# generate labels
# use Marital_Status
df_test = pd.DataFrame( data=df, columns=['Marital_Status'] )
df_test.head()
```

```python
pd.get_dummies(df_test)
```

```python
df_modelling_1 = df.drop('Dt_Customer', 1)
df_modelling_1
df_modelling = df_modelling_1.drop('Income', 1)
df_modelling
```

```python
df_modelling = pd.get_dummies(df_modelling)
df_modelling.head().transpose()
```

```python
# do train test split
# training set 70%, training machine learning models
# test set 30%, tset the model we have built to evaluate on the performance

# labels
x = df_modelling.drop('Response', 1)
y = df_modelling['Response']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.3, train_size=0.7 )
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# checking
X_train['ID'].count()
```

```python
X_test['ID'].count()
```

```python
# Build machine learning models: use random forest to build a ML model 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# defining the model to be used for training, 
clf = RandomForestClassifier(max_depth=15)
```

```python
clf.fit(X_train, y_train)
```

```python
type(clf) 
```

```python
# Make prediction on test dataset and test the model preformance
```

```python
y_predict = clf.predict(X_test)
y_predict
```

```python
# compare the y_predict, which is our prediction with the y_test, to evaluate on the accuracy of the modelling
from sklearn.metrics import accuracy_score, precision_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
# accuracy
accuracy_score(y_test, y_predict)
```

```python
# confusion matrix
confusion_matrix(y_test, y_predict)
```

```python
# precision score
precision_score(y_test, y_predict, average='macro')
```

```python
# AUC
y_prob = clf.predict_proba(X_test)
pred = y_prob[:,1]
fpr, tpr, thresholds = roc_curve(y_test, pred)
auc(fpr, tpr)
```

```python
# try to plot auc curve

# we assign the auc(fpr, tpr) to a value called roc_auc lrl
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, lw=2)
```

```python
plt.plot(fpr, tpr, lw=2)
plt.plot([0,1], [0,1], lw=2, linestyle='--')
```
