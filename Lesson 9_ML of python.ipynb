### For individual level, build a model to predict the probability of leaving current job

```python
# 4.1 do data preprocessing
# since data format can be very messy, not suitable for ML algorithms to directly consume
df.head()
```

```python
# take an example, why we need data poeprocessing, 
# e.g.BusinessTravel, it is a string, which is not numerical values,
# data poeprocessing can make them suitable to ML job
# then there also some desciptions of some cols (some individuals) of dataset, 

# For conducting ML, 
# we need labels, which is a tag, yes or no, which means whether leave this job or not
# labels must be numerical values if it is a classification problem, (0/1) , 
# it can be just a value if it is a regression

# make sure all values are numerical values
# for business travel, we do as below:
```

```python
# generate labels
```

```python
# one hot encoding, to convert all non-numerical features into numerical features 
# one hot encoding, it is called pandas.get_dummies
df_test = pd.DataFrame( data=df, columns=['BusinessTravel'] )
df_test.head()
```

```python
pd.get_dummies(df_test)
```

```python
# then, we transform other features to matrix form
```

```python
df_modelling = df.drop('Attrition', 1)
df_modelling.head()
```

```python
df_modelling = pd.get_dummies(df_modelling)
df_modelling.head().transpose()
```

```python
# do train test split
```

```python
df['Age'].count() # 1470
```

```python
# training set 70%, training machine learning models
# test set 30%, tset the model we have built to evaluate on the performance
```

```python
# labels
x = df_modelling.drop('Attrition_tag', 1)
y = df_modelling['Attrition_tag']
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

# train_test_split is to split one data frame to two data frame 
#                                                         according to a predefined ratio
```

```python
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.3, train_size=0.7 )
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test 
```

```python
# almost done -- data preprocessing
# x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.3, train_size=0.7 )
# is a data frame that we prepare for building model
```

```python
## Build machine learning models
```

```python
# here we use random forest to build a ML model that predict the employees' attrition
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# defining the model to be used for training, 
# *** max_depth= 3 or 5 for basic work, tesing ang changing values after 
clf = RandomForestClassifier(max_depth=3)
```

```python
# input the training features and labels to train the model
clf.fit(X_train, y_train)  # two parameters for this fit
```

```python
type(clf)  # clf is a model, but we cannot see what is inside now, wt info. inside
           #                 there is some complicated binary info. inside
```

```python
## Make prediction on test dataset and test the model preformance
```

```python
# model is ready
# we have X_test and y_test, but we ASSUME we only know the X_test
# then use it to predict the attrition of all these users,
#      compare with the y_test (the true label)
#      compare the accuracy and even auc to get uderstanding how accurate our model is
#                                             (the purpose of having this test set y_test 
#              --> to test the model performance assume that we dont know the exact label)
```

```python
# model is ready & compare with the y_test (the true label)
# do predictions, clf we have
y_predict = clf.predict(X_test)
```

```python
y_predict
```

```python
# compare the y_predict, which is our prediction with the y_test, 
# to evaluate on the accuracy of the modelling
# accuracy, confusion matrix, AUC
from sklearn.metrics import accuracy_score, precision_score, classification_report, confusion_matrix, roc_curve, auc
```

```python
# accuracy
accuracy_score(y_test, y_predict)
```

```python
# confusion matrix
confusion_matrix(y_test, y_predict)
```

```python
# precision score
precision_score(y_test, y_predict, average='macro')
```

```python
# AUC, 1 means perfect classification; 0.5 means random guess
```

```python
# first, go back to predict the prob.
y_prob = clf.predict_proba(X_test)
```

```python
y_prob  # y_prob shows the prob. being 0 or 1 while inputting to this roc curve, 
        # we only need the prob. being 1, i.e.2nd col
```

```python
pred = y_prob[:,1]
pred
```

```python
fpr, tpr, thresholds = roc_curve(y_test, pred)
auc(fpr, tpr)
```

```python
# 0.78 > 0.5, it is random guess, there is a room to increase it.
```

```python
# try to plot auc curve

# we assign the auc(fpr, tpr) to a value called roc_auc lrl
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, lw=2)
```

```python
# area under this curve is 1 means it is a prefect segmentation
plt.plot(fpr, tpr, lw=2)
plt.plot([0,1], [0,1], lw=2, linestyle='--')  # this is roc_auc 
```

```python
# this is a gd model because it is able to do some predictions 
# and it performs much better than random guess
# we can improve it much more better
# look at auc score (b/c optimize auc score = improve the predictions)
# we can go back to 'building machine learning model' part,
#                                                     try another max_depth no. e.g. 8
# do this training again and run
```
